```
folders = os.listdir(results)

folder = path.join(results, folders[1])

print(folder)
print(folders[1], path.exists(folder))

os.listdir(folder)
```

```
data = list()
for file in os.listdir(folder):
  index = file.rfind('_')
  technique = file[:index]
  print(technique)

  df = pd.read_csv(path.join(folder, file))
  df['feat_extraction'] = technique
  data.append(df)

dframe = pd.concat(data, axis=0, ignore_index=True)
del data
```
